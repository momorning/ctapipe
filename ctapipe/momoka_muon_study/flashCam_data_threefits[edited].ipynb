```python
from ctapipe.utils.datasets import get_dataset_path
from ctapipe.io import event_source, EventSeeker
from ctapipe.visualization import CameraDisplay
from ctapipe.image import tailcuts_clean
from ctapipe.image.muon import MuonRingFitter
from matplotlib import pyplot as plt
from astropy import units as u
import numpy as np
```

```python
filename1 = "/Users/momokagoto/ctaTestFiles/opt_eff_100/muon_0deg_0deg_run190___cta-prod3-mst-fc_desert-2150m-Paranal-mst-fc.simtel.gz"
```

```python
source1 = event_source(filename1)
for event in source1:
    
    if event.count >100:
        break
    
    from ctapipe.calib import CameraCalibrator
    calib = CameraCalibrator()
    calib(event)

    tel_id = sorted(event.r1.tels_with_data)[0]
    sub = event.inst.subarray
    camera = sub.tel[tel_id].camera
    optics = sub.tel[tel_id].optics
    image = event.dl1.tel[tel_id].image

    for telid in event.dl0.tels_with_data:

        # Get geometry
        teldes = event.inst.subarray.tel[telid]
        focal_length= teldes.optics.equivalent_focal_length
        print("optics", teldes.optics.equivalent_focal_length)
        print("camera", teldes.camera)
        geom = teldes.camera

        mask = tailcuts_clean(geom, image, 10, 12)
        image[~mask] = 0
        x, y = np.array(geom.pix_x), np.array(geom.pix_y)
        
        squared_distanced = (x[0] - x[1:])**2 + (y[0] - y[1:])**2
        angle_pix_width = np.sqrt(squared_distanced.min())
        
        for method in MuonRingFitter.fit_method.values:

            muonfit = MuonRingFitter(fit_method=method)
            fit_result = muonfit(geom.pix_x, geom.pix_y, image, mask)
            print(method,fit_result)
            
            #plt.figure(figsize=(5,5))
            disp = CameraDisplay(geom, image)
            
            xc_fit = fit_result.ring_center_x.to_value(u.m) 
            yc_fit = fit_result.ring_center_y.to_value(u.m) 
            r_fit = fit_result.ring_radius.to_value(u.m)  
            centroid= (xc_fit,yc_fit)
            dist= np.sqrt((geom.pix_x.value-xc_fit)**2 + (geom.pix_y.value-yc_fit)**2)
            ring_dist= np.abs(dist - r_fit)
            cut_outlier = ring_dist < r_fit * 0.4
            combined_mask = mask & cut_outlier
            chisquare = (np.sum(((ring_dist[combined_mask]/(angle_pix_width))**2))/(sum(combined_mask)))
            if method == "kundu_chaudhuri":
                #disp.add_ellipse(centroid = centroid, length=r_fit*2, width=r_fit*2, angle=0,color = 'yellow',lw = 2)
                allradius_ck.append(r_fit)
                allcenterx_ck.append(xc_fit)
                allcentery_ck.append(yc_fit)
                chisquares_ck.append(chisquare)
            elif method == "taubin":
                #disp.add_ellipse(centroid = centroid, length=r_fit*2, width=r_fit*2, angle=0,color = 'red',lw = 2)
                allradius_t.append(r_fit)
                allcenterx_t.append(xc_fit)
                allcentery_t.append(yc_fit)
                chisquares_t.append(chisquare)
            elif method == "hough":
                #disp.add_ellipse(centroid = centroid, length=r_fit*2, width=r_fit*2, angle=0,color = 'blue',lw = 2)
                allradius_h.append(r_fit)
                allcenterx_h.append(xc_fit)
                allcentery_h.append(yc_fit)
                chisquares_h.append(chisquare)
        pixelcharge = sum(image[mask])
        pixelcharge_all.append(pixelcharge)
        pixelnum = len(image[mask])
        pixelnumber_all.append(pixelnum)
        avg_dist = sum(np.sqrt((np.array(x)[mask])**2 + (np.array(y)[mask])**2))

        avgdist_list.append(avg_dist/len(image[mask]))
```
