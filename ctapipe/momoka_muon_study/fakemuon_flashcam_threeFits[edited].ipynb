```python
from ctapipe.image import toymodel
from ctapipe.instrument import CameraGeometry
from ctapipe.io import event_source, EventSeeker
from ctapipe.visualization import CameraDisplay
from ctapipe.image import tailcuts_clean
from ctapipe.image.muon import MuonRingFitter
from matplotlib import pyplot as plt
from astropy import units as u
import numpy as np
```

```python
fit_values=[]
true_inclination_list=[]
true_radius_list=[]
true_centerx_list=[]
true_centery_list=[]
pixelcharge_all=[]
pixelnumber_all=[]
avgdist_all=[]


allcenterx_t=[]
allcentery_t=[]
allradius_t=[]
error_xvalues_t=[]
error_yvalues_t=[]
error_radius_t=[]
chisquares_t=[]
inclination_list_t=[]

allcenterx_ck=[]
allcentery_ck=[]
allradius_ck=[]
error_xvalues_ck=[]
error_yvalues_ck=[]
error_radius_ck=[]
chisquares_ck=[]
inclination_list_ck=[]

allcenterx_h = []
allcentery_h = []
allradius_h = []
error_xvalues_h = []
error_yvalues_h = []
error_radius_h = []
chisquares_h = []
inclination_list_h = []
```

```python
for n in range(100):
    num_tests = 1
    center_xs = np.random.uniform(-0.8, 0.8)
    center_ys = np.random.uniform(-0.8, 0.8)
    ring_radius = np.random.uniform(0.2, 0.5)#1/7 of size of the camera
    ring_width = np.random.uniform(0.04,0.08) #1/5 of radius
    true_inclination = np.sqrt((center_xs)**2 + (center_ys)**2)
    true_inclination_list.append(true_inclination)
    true_centerx_list.append(center_xs)
    true_centery_list.append(center_ys)
    true_radius_list.append(ring_radius)
    geom=CameraGeometry.from_name("FlashCam")
    focal_length = u.Quantity(16, u.m)
    muon_model = toymodel.RingGaussian(x=center_xs* u.m, 
                                       y=center_ys* u.m,
                                       radius=ring_radius* u.m, 
                                       sigma=ring_width* u.m
    )
    image, sig, bg = muon_model.generate_image(
            geom, intensity=1000, nsb_level_pe=5,
        )
    
    mask = tailcuts_clean(geom, image, 10, 12)
    image[~mask] = 0
  
    x, y = np.array(geom.pix_x), np.array(geom.pix_y)

    squared_distanced = (x[0] - x[1:])**2 + (y[0] - y[1:])**2
    angle_pix_width = np.sqrt(squared_distanced.min())#*u.deg

    for method in MuonRingFitter.fit_method.values:

        muonfit = MuonRingFitter(fit_method=method)
        fit_result = muonfit(geom.pix_x, geom.pix_y, image, mask)
        print(method,fit_result)

        #plt.figure(figsize=(5,5))
        disp = CameraDisplay(geom, image)

        xc_fit = fit_result.ring_center_x.to_value(u.m) 
        yc_fit = fit_result.ring_center_y.to_value(u.m) 
        r_fit = fit_result.ring_radius.to_value(u.m)  
        centroid= (xc_fit,yc_fit)
        dist= np.sqrt((geom.pix_x.value-xc_fit)**2 + (geom.pix_y.value-yc_fit)**2)
        ring_dist= np.abs(dist - r_fit)
        cut_outlier = ring_dist < r_fit * 0.4
        combined_mask = mask & cut_outlier
        chisquare = (np.sum(((ring_dist[combined_mask]/(angle_pix_width))**2))/(sum(combined_mask)))
        error_radius=(xc_fit - ring_radius)/ring_radius
        error_xvalues=(yc_fit - center_xs)/center_xs
        error_yvalues=(r_fit - center_ys)/center_ys
        if method == "kundu_chaudhuri":
            #disp.add_ellipse(centroid = centroid, length=r_fit*2, width=r_fit*2, angle=0,color = 'yellow',lw = 2)
            allradius_ck.append(r_fit)
            allcenterx_ck.append(xc_fit)
            allcentery_ck.append(yc_fit)
            chisquares_ck.append(chisquare)
            error_radius_ck.append(error_radius)
            error_xvalues_ck.append(error_xvalues)
            error_yvalues_ck.append(error_yvalues)
        elif method == "taubin":
            #disp.add_ellipse(centroid = centroid, length=r_fit*2, width=r_fit*2, angle=0,color = 'red',lw = 2)
            allradius_t.append(r_fit)
            allcenterx_t.append(xc_fit)
            allcentery_t.append(yc_fit)
            chisquares_t.append(chisquare)
            error_radius_t.append(error_radius)
            error_xvalues_t.append(error_xvalues)
            error_yvalues_t.append(error_yvalues)
        elif method == "hough":
            #disp.add_ellipse(centroid = centroid, length=r_fit*2, width=r_fit*2, angle=0,color = 'blue',lw = 2)
            allradius_h.append(r_fit)
            allcenterx_h.append(xc_fit)
            allcentery_h.append(yc_fit)
            chisquares_h.append(chisquare)
            error_radius_h.append(error_radius)
            error_xvalues_h.append(error_xvalues)
            error_yvalues_h.append(error_yvalues)
    pixelcharge = sum(image[mask])
    pixelcharge_all.append(pixelcharge)
    pixelnum = len(image[mask])
    pixelnumber_all.append(pixelnum)
    avg_dist = sum(np.sqrt((np.array(x)[mask])**2 + (np.array(y)[mask])**2))
    avgdist_all.append(avg_dist/(len(image[mask])))
```

```python
plt.hist(chisquares_h,label="h")
plt.hist(chisquares_t,label="t")
plt.hist(chisquares_ck,label="ck")
plt.legend()
plt.xlabel("chisquares")
plt.ylabel("Counts")
plt.title("FlashCam fakemuon file:Chisquares with ThreeFits")
```

```python
plt.plot(pixelcharge_all,chisquares_h,"yo",label="h")
plt.plot(pixelcharge_all,chisquares_ck,"ro",label="ck")
plt.plot(pixelcharge_all,chisquares_t,"go",label="t")
plt.legend()
plt.xlabel("pixel charge")
plt.ylabel("chisquare")
plt.title("Fakemuon FlashCam PixelCharge/Chisquare")
```

```python
plt.plot(pixelnumber_all,chisquares_h,"yo",label="h")
plt.plot(pixelnumber_all,chisquares_ck,"ro",label="ck")
plt.plot(pixelnumber_all,chisquares_t,"go",label="t")
plt.legend()
plt.xlabel("pixel number")
plt.ylabel("chisquare")
plt.title("Fakemuon FlashCam PixelNum/Chisquare")
```

```python
plt.plot(avgdist_all,chisquares_ck,"ro",label="ck")
plt.plot(avgdist_all,chisquares_h,"yo",label="h")
plt.plot(avgdist_all,chisquares_t,"go",label="t")
plt.legend()
plt.xlabel("avgdist")
plt.ylabel("chisquare")
plt.title("Fakemuon FlashCam Three Fits: Avgdist vs Chisquare")
```
